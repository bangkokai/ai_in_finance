```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
data = pd.read_excel('คำตอบ.xlsx')
```

```python
print(data['1.อาชีพ'].value_counts())
data['1.อาชีพ'].value_counts().plot(kind="bar")
plt.title('Occupation')
```

```python
print(data['2.อายุ'].value_counts())
data['2.อายุ'].value_counts().plot(kind="bar")
plt.title('Age')
```

```python
hf_dict = {}
hf_dict['E-mail'] = 0
hf_dict['Facebook'] = 0
hf_dict['Friends'] = 0
hf_dict['Line'] = 0
hf_dict['Eventbrite'] = 0

hf_list = data['3.คุณทราบข่าวการจัดงานในวันนี้จากช่องทางใด'].tolist()
for hf in hf_list: 
    if 'mail' in hf:
        hf_dict['E-mail'] += 1
    if 'Facebook' in hf:
        hf_dict['Facebook'] += 1
    if ('เพื่อน' in hf) or ('ถูกพามา' in hf):
        hf_dict['Friends'] += 1
    if 'Line' in hf:
        hf_dict['Line'] += 1
    if 'brite' in hf:
        hf_dict['Eventbrite'] += 1
print(hf_dict)

pd.DataFrame.from_dict(hf_dict, orient='index').plot(kind="bar")
print('คุณทราบข่าวการจัดงานในวันนี้จากช่องทางใด')
```

```python
k_dict = {}
k_dict['No'] = 0
k_dict['Yes'] = 0

k_list = data['6.คุณเคยรู้จัก BANGKOK AI Community มาก่อนหน้านี้หรือไม่-อย่างไรบ้าง'].replace(np.nan, '', regex=True).tolist()
for k in k_list: 
    if ('ไม่' in k) or ('No' in k) or ('no' in k):
        k_dict['No'] += 1
    elif ('รู้' in k) or ('Yes' in k) or ('เคย' in k):
        k_dict['Yes'] += 1
print(k_dict)

pd.DataFrame.from_dict(k_dict, orient='index').plot(kind="bar")
print('เคยรู้จัก BANGKOK AI Community มาก่อนหน้านี้หรือไม่')
```
